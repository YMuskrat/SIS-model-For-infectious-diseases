```python
# Code for task T2 of Unit 5 Lab class 1
import numpy as np
import matplotlib.pyplot as plt
# Some f function
def f(x, R):
    return 1000*(1-(1/R)) # logistic map with parameter r >= 1
        
def bifurcation(r, iter):
    # One initial condition for each value of r
    x = np.random.rand(len(r))
    # We are first going to get rid of the transients
    for i in range(5000):
        x = f(x,r)
    # Assuming we are now on the limit cycle, we plot value of iter outputs
    for i in range(iter): 
        x = f(x,r)
        ax = fig.add_subplot(111)
        ax.plot(r,x, '.', color = 'k', markersize = 1)
        ax.set_ylim(0.0)
        
# First plot the function

# Bifurcation plot
fig = plt.figure(figsize=(8, 6), dpi=80)


R = np.arange(0.1,5.0,0.2)
iter = 1000
bifurcation(R, iter)
plt.show()
```







> Using the integrated B we achieved, which is \begin{align}
        B(t) = \frac{B^*}{(B^*/B -1){e}^{-t*beta*B^*/N}+1}
    \end{align}



> We plot B(t) over time, while changing R0 and we start by a population of 1 at time t=0 in B. We should note that, our system can behave if our B at time t=0 is less than 1.






> # When B0=1



```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

# function that returns dy/dt
def model(R,gamma,N,B0,t):
    Bstar=(1-1/R)*N
    exp=np.exp(-R*gamma*Bstar*t/N)
    Bt=Bstar/((Bstar/B0 -1)*exp +1)
    return Bt



# time points
t = np.linspace(0,20)


# R0<1
y1=model(0.1,0.5,1000,1,t)
y2=model(0.5,0.5,1000,1,t)
y3=model(0.9,0.5,1000,1,t)

#R0>1
y4=model(2,0.5,1000,1,t)
y5=model(2.5,0.5,1000,1,t)
y6=model(3,0.5,1000,1,t)
y7=model(5,0.5,1000,1,t)
# plot results
figure(figsize=(8, 6), dpi=80)
plt.plot(t,y1,'r-',linewidth=2,label='R0=0.1')
plt.plot(t,y2,'b--',linewidth=2,label='R0=0.5')
plt.plot(t,y3,'g:',linewidth=2,label='R0=0.9')

plt.plot(t,y4,'y-',linewidth=2,label='R0=1.7')
plt.plot(t,y5,'m--',linewidth=2,label='R0=2.5')
plt.plot(t,y6,'c:',linewidth=2,label='R0=3')

plt.plot(t,y6,'k:',linewidth=2,label='R0=5')



plt.title('B(t) over time')
plt.xlabel('time')
plt.ylabel('y(t)')
plt.legend(loc='best')
plt.show()
```

# **b) Euler Method**



> In the following Code, we adopted the numpy Euler function to integrate our function and see if our derived B(t) is correct and they converge to the same values. We mimmic the same parameter values we did when we used B(t), and we plot our graph, here we try for B0=1 and B0=500, while varrying R0.





> # When B0=1



```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dy/dt
def model(x,t,R,gamma,N):
    dxdt = R*gamma*x*(N-x)/N-gamma*x
    return dxdt

# initial condition
x = 1

# time points
t = np.linspace(0,20)

# solve ODEs
#R0<1
y1 = odeint(model,x,t,args=(0.1,0.5,1000))
y2 = odeint(model,x,t,args=(0.5,0.5,1000))
y3 = odeint(model,x,t,args=(0.9,0.5,1000))

#R0>1
y4 = odeint(model,x,t,args=(2,0.5,1000))
y5 = odeint(model,x,t,args=(2.5,0.5,1000))
y6 = odeint(model,x,t,args=(3,0.5,1000))
y7 = odeint(model,x,t,args=(5,0.5,1000))


figure(figsize=(8, 6), dpi=80)
plt.plot(t,y1,'r-',linewidth=2,label='R0=0.1')
plt.plot(t,y2,'b--',linewidth=2,label='R0=0.5')
plt.plot(t,y3,'g:',linewidth=2,label='R0=0.9')

plt.plot(t,y4,'y-',linewidth=2,label='R0=1.7')
plt.plot(t,y5,'m--',linewidth=2,label='R0=2.5')
plt.plot(t,y6,'c:',linewidth=2,label='R0=3')

plt.plot(t,y6,'k:',linewidth=2,label='R0=5')



plt.title('B(t) over time using Euler')
plt.xlabel('time')
plt.ylabel('y(t)')
plt.legend(loc='best')
plt.show()
```



> When R0=1 we get a straight line which devides the plotted lines when R0>1 and when R0<1 , we can say that R0 act as the treshhold of our system, for it to either decrease or increase overtime. In other words, if R0 pass that treshhold which is 1 our system behavior increases, and vice versa.



```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dy/dt
def model(x,t,R,gamma,N):
    dxdt = R*gamma*x*(N-x)/N-gamma*x
    return dxdt

# initial condition
x = 1

y1 = odeint(model,x,t,args=(0.5,1,1000))
y2 = odeint(model,x,t,args=(0.5,0.7,1000))
y3 = odeint(model,x,t,args=(2,10,1000))
y4 = odeint(model,x,t,args=(2.1,20,1000))

figure(figsize=(8, 6), dpi=80)
plt.plot(t,y1,'r-',linewidth=2,label='gamma=0.1')
plt.plot(t,y2,'b',linewidth=2,label='gamma=0.7')
plt.plot(t,y3,'k',linewidth=2,label='gamma=10')
plt.plot(t,y4,'y',linewidth=2,label='gamma=20')

plt.title('B(t) over time when we varry Gamma')
plt.xlabel('time')
plt.ylabel('y(t)')
plt.legend(loc='best')
plt.show()
```



> When we increase the value of gamma we can see the convergence of our graphs is much quicker than when gamma is smaller. the bigger gamma the quicker the convergence of our system. if our R0>1 our system converges to 500+, when our R0<1 our system converges to 0 much quicker.
